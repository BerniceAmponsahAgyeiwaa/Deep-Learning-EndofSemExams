```python
import pandas as pd

results = pd.DataFrame([
    {
        "model": "Baseline LSTM",
        "accuracy": 0.2889,
        "f1": 0.1295,
        "fairness_level": "Low"
    },
    {
        "model": "Transformer + SVM (Improved)",
        "accuracy": 0.6593,
        "f1": 0.6589,
        "fairness_level": "Medium"
    },
    {
        "model": "Adversarial LSTM (Debiased)",
        "accuracy": 0.2857,
        "f1": 0.1270,
        "fairness_level": "High"
    }
])

results
```

```python
# save results
results.to_csv("../results/final_model_comparison.csv", index=False)
```

```python
# accuracy fairness tradeness plot
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.scatterplot(
    data=results,
    x="accuracy",
    y="f1",
    hue="model",
    s=200
)
plt.title("Accuracy–Fairness Trade‑off Across Models")
plt.xlabel("Accuracy")
plt.ylabel("F1 Score")
plt.grid(True)
plt.savefig("../results/accuracy_fairness_tradeoff.png", dpi=300, bbox_inches="tight")
plt.show()
```
